```python
soup=BeautifulSoup(web.content,"html.parser")
```

```python
print(soup.prettify())
```

```python
class_data = soup.find("div",class_ ="td-post-content")
```

```python
class_data
```

```python
lines=class_data.find_all(["p", "ol","ul","h1","h3","a"])
filtered_data = set()
for line in lines:
    text = line.text.strip()
    filtered_data = list(filtered_data)
    for item in filtered_data:
        print(item)
```

```python
import csv
from bs4 import BeautifulSoup

# Assuming class_data is your BeautifulSoup object
lines = class_data.find_all(["p", "ol", "ul", "h1", "h3", "a","li"])

# Extract text data from the elements
text_data = [line.text.strip() for line in lines]

# Specify the CSV file name
csv_file = "scraped_data.csv"

# Write the text data to the CSV file
with open(csv_file, "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Text"])  # Writing the header
    for text in text_data:
        writer.writerow([text])  # Writing each line of text

print(f"Data has been saved to {csv_file}")
```
